## **DICTIONARY**

1. It is a built in data type that stores data as key-value pairs
2. Keys are unique & immutable(eg. strings)
3. Implementation={}
4. Dictionary is mutuable


```python
student={"Name":"Ishant","Roll-no.":24,"Marks":85}
print(student,type(student),sep='\n')
```

```python
student['Marks']=90 #updates the value of the key
print(student)

student['Section']="A1" #if key is not present then it appends it in the dict
print(student)
```

```python
for i,j in  student.items():
    print(i,": ",j)
```

```python
print(student.values())
```

```python
for i in student.items():
    print(i[1])
```

```python
print(student.keys())
```

```python
print(student.get('Name'))#if we want to print the item of the key

print(student.get('Course',12))#if the key is not present in dict then it will return other value
```

```python
student.update({"Marks":80}) #can update more than one values
print(student)

del student["Marks"] #to delete the key
print(student)
```
