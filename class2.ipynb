```python
##SEQUENTIAL DATA TYPE:-
#List
#Tuple
#Range

'''Mutable- Integers are immutable as when we change the value to different data
type then it will create a new memory and then will point to new memory
eg. a=10
    a=12.2 '''

ar=[1,2,3,4,"A"] #List- can store homogenous as well as heterogenous type of data types
print(ar,type(ar))#List are mutuable
ar.append("abc")#repeated element can be there
print(ar,type(ar)) #data added
print(ar[2])#order of insertion is preserved
```

```python
arr=(1,2,3,4,"A") #Tuple are immutable
print(arr,type(arr)) #Tuple-can store homogenous as well as heterogenous type of data types

arr2=("a","b","c")#repeated element can be there
arr2=arr+arr2 #Addition is possible
print(arr2,type(arr))
```

```python
#Range
value=range(1,11,2)#It is a data type
print(value,type(value))
for i in value:#it goes upto n-1
    print(i)

for i in range(11,1,-1):
    print(i)
```

```python
#NON-SEQUENTIAL:- 
"""SET DATATYPE"""
ar2={1,2,3,4,4,4,4,'5'}#it does't contain duplicate element
print(ar2,type(ar2))
ar2.add('abc')#mutuable
#print(ar2[4])##order of insertion is not preserved
ar2.remove(1)#it can remove the value directly from the set
print(ar2)
```

```python
#Dictionary-Non Sequential data type
ar3={1:'a',2:'b',3:'c',3:'d','4':'d'}#it cannot store the duplicate key but will update itself to the newest value
print(ar3,type(ar3))#value can be duplicate
print(ar3[1])
print(ar3['4'])
ar3.update({1:'z'})
print(ar3)
ar3.update({5:'e'})#mutable
print(ar3)
```
