## **Custom Modules**
A custom module is a Python file (.py) that you create to organize your code. It lets you group related functions, classes, and variables into a single, reusable unit.

How to Use a Custom Module
Using a custom module is a simple, two-step process: you create the module and then import it into another script or notebook.

**In one script(script.py)**

```python
def say_hello(name):
    return f"Hello,{name}!"
```

**In other script named script.py<br>
can be in any other(but we have to mention the path)**

```python
from module import script

print(script.say_hello("Ishant")) #these will not run in jupiter notebook
```
