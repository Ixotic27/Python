```python
#INHERITANCE	
#Its one of the features of OOPS.By using inheritance we can reuse the existing code (class).	
#We can create multiple child classes of single super class.

class Parent:
    def add(self,a,b):
        return a+b
    
class Child1(Parent):
    def displayAdd(self,a,b):
        print(super().add(a,b))

obj=Child1()
obj.displayAdd(2,3)
#By using inheritance we can acess the visible attributes and features of parent class.
```

# **TYPES OF INHERITANCE**<br>
1. Single Level<br>
2. Multilevel Inheritance<br>
3. Heirarchy Inheritance<br>
4. Hybrid Inheritance<br>

```python
#SINGLE LEVEL INHERITENCE
class Parent:
    c=10
    def add(self,a,b):
        return a+b
class Child1(Parent):
    def display(self,a,b):
        print(super().add(a,b))
        print(super().c)

obj =Child1()
obj.display(12,12)
```

```python
# Hybrid Inheritance Example - It is the combination of any of the type of inheritence

class Creature:  # Base Class
    def __init__(self, name):
        self.name = name

    def breathe(self):
        return f"{self.name} breathes."

class Animal(Creature):  # Derived from Creature (hierarchical branch 1)
    def move(self):
        return f"{self.name} moves on land."

class Plant(Creature):  # Derived from Creature (hierarchical branch 2)
    def grow(self):
        return f"{self.name} grows through photosynthesis."

class Mammal(Animal):  # Multilevel inheritance (extends Animal)
    def feed_milk(self):
        return f"{self.name} feeds milk to its young ones."

class Human(Mammal):  # Further multilevel
    def speak(self):
        return f"{self.name} can speak and reason."

# Usage
h = Human("Human")
p = Plant("Fern")

print(h.breathe())      # inherited from Creature
print(h.move())         # inherited from Animal
print(h.feed_milk())    # from Mammal
print(h.speak())        # from Human
print(p.grow())         # from Plant

# Output:
# Human breathes.
# Human moves on land.
# Human feeds milk to its young ones.
# Human can speak and reason.
# Fern grows through photosynthesis.


'''        Creature             ← Base Class
          /      \
     Animal      Plant          ← Hierarchical (two derived)
       |
     Mammal                     ← Multilevel (extends Animal)
       |
     Human                      ← Multilevel continuation
'''
```

# **Method Overiding**<br>
In Python, method overriding is an object-oriented programming concept that allows a subclass to provide a specific implementation of a method that is already provided by its parent class. When a method in a subclass has the same name, parameters, and return type as a method in its superclass, the method in the subclass is said to override the method in the superclass.

```python
# 1. Parent Class (Superclass)
class Animal:
    """A general representation of an animal."""
    def __init__(self, name):
        self.name = name

    def speak(self):
        """Generic sound for an animal."""
        return f"{self.name} makes a generic animal sound."

# 2. Child Class (Subclass) - Dog
class Dog(Animal):
    """A dog that inherits from Animal."""
    def speak(self):
        """Overrides the parent's speak method for a dog."""
        return f"{self.name} says Woof!"

# 3. Child Class (Subclass) - Cat
class Cat(Animal):
    """A cat that inherits from Animal."""
    def speak(self):
        """Overrides the parent's speak method for a cat."""
        return f"{self.name} says Meow!"

# --- Demonstration ---

def make_animal_speak(Animal): #Unified data base
    print(Animal.speak())

# Create instances of each class
generic_animal = Animal("Creature")
dog = Dog("Rex")
cat = Cat("Whiskers")

# Call the 'speak' method on each object.
# Notice how the same method call behaves differently for each object.
make_animal_speak(generic_animal)
print(generic_animal.speak())
print(dog.speak())
print(cat.speak())
```

# **Polymorphism in python using unified interface**<br>
Polymorphims in Python,particulary through a "unified interface:, refers to the ability to treat objects of different classes in a consistent manner, as kibg as they share a common set of methods or attrivutes. This is often achieved through inheritence or by simply ensuring that different classes implement the same method names(duck typing).

```python
# Base class (optional)
class Animal:
    def make_sound(self):
        pass  # acts as a placeholder


# Derived classes implementing the same interface
class Dog(Animal):
    def make_sound(self):
        return "Bark!"

class Cat(Animal):
    def make_sound(self):
        return "Meow!"

class Cow(Animal):
    def make_sound(self):
        return "Moo!"


# Unified interface function
def animal_sound(animal):
    # Same interface for all derived classes
    print(animal.make_sound())


# Polymorphic behavior
animals = [Dog(), Cat(), Cow()]

for a in animals:
    animal_sound(a)
```

# **Encapsulation**<br>
1. Encapsulation = data hiding + controlled access<br>
2. In OOP, encapsulation means bundling data (variables) and methods (functions) that operate on that data into a single unit (class), and restricting direct access to some of the object’s components.

```python
class Student:
    def __init__(self, name, roll_no, marks):
        self.name = name          # public attribute
        self.roll_no = roll_no    # public attribute
        self.__marks = marks      # private attribute (encapsulated)
    
    # Getter method (to access private variable)
    def get_marks(self):
        return self.__marks
    
    # Setter method (to modify private variable safely)
    def set_marks(self, marks):
        if 0 <= marks <= 100:
            self.__marks = marks
        else:
            print("❌ Invalid marks! Must be between 0 and 100.")
    
    # Display method
    def show_details(self):
        print(f"Marks: {self.__marks}")

s= Student ("Ishsnt",24,85)
print(s.name)
print(s.roll_no)
# print(s.__marks) Its an encapsulated/private member which cannot be accessed without the class method as it cannot be direclty accessed
#outside the class
s.show_details() #It can be accesde only through class method
```
