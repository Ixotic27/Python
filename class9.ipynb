OBJECT ORIENTED PROGRAMMING<br>

OOP is a programming paradigm that organizes code around "objects" rather than functions and logic. Think of it as creating blueprints (classes) for things, and then from those blueprints, you create actual instances (objects). This approach helps in managing complexity, promoting code reuse, and making your programs more organized.

This is the most fundamental concept in OOP.

A Class is like a blueprint. For example, a Car blueprint would define general properties like color and max_speed, and behaviors like start_engine() and accelerate().

An Object is the actual thing created from that blueprint. So, a red Ferrari or a blue Tesla would be two different objects (or instances) of the Car class.

## Creating a Class
The sum() method is a special method called a constructor. It's automatically called when you create a new object from the class. The self parameter refers to the specific instance of the class being created.

```python
#Class - It is a combination of attributes and methods and such methods always share similar attributes 
class Data:
    c=10#Attributes
    def sum(self,a,b): #Method
        return a+b
obj=Data()
print(obj.sum(5,5))
```

```python
class calculator:
    c=10
    def __init__(self,a,b):
        self.a=a
        self.b=b
    
    def sum(self):
        return self.a+self.b
    def sub(self):
        return self.a-self.b
    def smul(self):
        return self.a*self.b
    def div(self):
        return self.a/self.b
    
obj=calculator(1,2)
print(obj.sum())
print(obj.c)
```
