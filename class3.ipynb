````python
import math as mt
def prime(n):
    if n<=1:
        print("Not Prime!")
        return
    found=1
    r=int(mt.sqrt(n))
    for i in range(2,r,1):
       if n%i==0:
          found-=1
          break
    if(found): print("Prime no")
    else : print("Not a prime no")
         
a=int(input("Enter the value of a: "))
prime(a)
````

# ðŸ“¦ Python Modules: An Introduction

A **module** in Python is simply a file containing Python code. This file can define functions, classes, and variables. Modules are a fundamental way to organize code into reusable units, helping to keep your programs clean and manageable.

---

## 1. Built-in Modules

**Built-in modules** are a set of pre-installed, standard libraries that come with the Python interpreter. They provide a wide range of functionalities without needing any external installation. 

### How to Use Built-in Modules

To use the code from a built-in module, you must first **import** it.

#### A. Importing the Entire Module
This is the most common method. You access its contents using the dot (`.`) operator.

```python
# Import the entire 'math' module
import math

# Use the 'math' module to access its functions
print(math.sqrt(16))  # Output: 4.0
print(math.pi)      # Output: 3.141592653589793

````python
# Import the 'math' module and assign it the alias 'mt'
# This is a common practice to improve readability and reduce typing
import math as mt

# Now, you can use 'mt' to access all the functions and constants
# from the 'math' module, like sqrt() and pi.

# Call the square root function using the alias 'mt'
print(mt.sqrt(16))

# Access the constant 'pi' using the alias 'mt'
print(mt.pi)
````
