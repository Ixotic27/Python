## **SET**
1. Are mutable data types as we can add elements in it
2. they donot contain duplicate element

```python
l1=[1,2,3,3,3,4,4]
s1=set(l1)
print(s1,type(s1))
```

```python
l1=list(s1)
print(l1)
```

```python
s1.add(5) #can add single value
print(s1)

s1.update({6,7,8}) #can add multiple value
print(s1)
```

```python
try:
    s1.remove(7)
except KeyError:
    print("Done!")
s1.discard(6)
print(s1)
```

```python
A={1,2,3}
B={3,4,5}

print(A.union(B))
print(A.intersection(B))
print(A.difference(B)) #A-B(Elements present in A only)
print(A.symmetric_difference(B)) #A' intersection B'
```

```python
print(A|B)
print(A&B)
print(A-B)
print(A^B)
```
