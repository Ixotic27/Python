## **FILE HANDLING**
Modes you can use:

"r" â†’ read (default, file must exist)

"w" â†’ write (creates file, erases old content)

"a" â†’ append (adds new content at end)

"x" â†’ create (fails if file already exists)

"b" â†’ binary mode (like "rb", "wb")

"t" â†’ text mode (default)

```python
#to open a file
f = open("myfile.txt", "w")  # open in write mode
```

```python
#Writing in a file
f.write("Hello, this is my first file!\n")
f.write("Python is fun.\n")
f.write("Made by Ishant Singh Bisht\n")
f.close()
```

```python
#Reading in a file
f = open("myfile.txt", "r")
#content = f.read()
#print(content)
print(f.read())
f.close()
```

```python
#Reading One line or Multiple Lines
f = open("myfile.txt", "r")

print(f.readline())      # reads one line
print(f.readlines())     # reads all lines into a list

f.close()
```

```python
#Appending to a File
f = open("myfile.txt", "a")
f.write("Adding a new line at the end.\n")
f.close()

f=open("myfile.txt",'r')
print(f.read())
f.close()
```

```python
#5. Using with (Best Practice âœ…)

#This automatically closes the file for you:

with open("myfile.txt", "r") as f:
    for line in f:
        print("->", line.strip(), "<-",)
  # strip removes newline
```

```python
#Different Reading Methods

#Iterating by characters
with open("myfile.txt", "r") as f:
    char = f.read(5)   # read first 5 characters
    print(char)
```

```python
#Writing Lists of Lines
lines = ["Line 1\n", "Line 2\n", "Line 3\n"]

with open("myfile.txt", "w") as f:
    f.writelines(lines)


#âš ï¸ Notice you must add \n yourself, unlike print().
```

```python
#3. File Pointer (seek and tell)

#Every file has a cursor/pointer that moves as you read/write.

with open("myfile.txt", "r") as f:
    print(f.tell())       # current position
    f.read(5)             # read 5 characters
    print(f.tell())       # new position
    f.seek(0)             # move cursor to beginning
    print(f.read(5))      # read again from start


#ðŸ‘‰ tell() = where am I?
#ðŸ‘‰ seek(pos) = move cursor to position.
```

```python
#4. Working with Binary Files

#For images, PDFs, etc., use binary mode "rb" / "wb".

# Copy an image
# To run the below code make sure you have an image name "source.jpg" should be present
# in your current working directory
with open("source.jpg", "rb") as src, open("copy.jpg", "wb") as dst:
    dst.write(src.read())
```

```python
#5. Exception Handling in File Operations

#Itâ€™s common for files to be missing or locked. Use try-except.

try:
    with open("nofile.txt", "r") as f:
        print(f.read())
except FileNotFoundError:
    print("File does not exist!")
```

```python
#6. File & Directory Management with os
import os

print(os.getcwd())   # current directory
os.mkdir("new_folder")    # make folder
os.rename("old.txt", "new.txt")   # rename file
os.remove("new.txt")      # delete file
os.rmdir("new_folder")    # remove empty folder
```

```python
#7. Working with with open in Multiple Files
with open("myfile.txt", "r") as infile, open("output.txt", "w") as outfile:
    for line in infile:
        outfile.write(line.upper())
with open("output.txt",'r') as f:
    print(f.read())

#ðŸ‘‰ Reads from one file, writes uppercase version into another.
```

```python
#8. JSON Files (Very Common in ML/Web Projects)
import json

data = {"name": "Ishant", "age": 20, "skills": ["Python", "ML"]}

# Write JSON
with open("data.json", "w") as f:
    json.dump(data, f)

# Read JSON
with open("data.json", "r") as f:
    loaded = json.load(f)
    print(loaded["name"])
    print(loaded)
```

```python
#9. CSV Without Pandas
import csv

# Write
with open("people.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Name", "Age"])
    writer.writerow(["Ishant", 20])
    writer.writerow(["Devendra", 21])

# Read
with open("people.csv", "r") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
```
